```python
from gensim.models import Word2Vec

# 示例语料
sentences = [
    ["cat", "jumps", "over", "the", "mouse"],
    ["king", "wears", "a", "golden", "crown"]
]

# 训练Skip-Gram模型
model = Word2Vec(
    sentences,
    vector_size=100,    # 词向量维度
    window=5,           # 窗口大小
    sg=1,               # 1表示Skip-Gram，0表示CBOW
    negative=5,         # 负采样数量
    min_count=1         # 忽略低频词
)

# 获取“cat”的词向量
vector = model.wv["cat"]

# 找到最相似的词
similar_words = model.wv.most_similar("cat", topn=3)
```
